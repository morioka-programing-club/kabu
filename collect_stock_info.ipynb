```python
import pandas as pd
import numpy as np
import jsm
import datetime
from tqdm import tqdm
from os.path import expanduser

now = datetime.date.today()#日付を取得
q=jsm.Quotes()#インスタンスを作成
c=q.get_brand()#全銘柄の、銘柄コード・市場・会社名・情報を取得
```

```python
code_list=pd.DataFrame({"ccode":[],"market":[],"name":[],"info":[]})#取得した銘柄情報のを入れるリストを用意

n=0
for i in tqdm(c.keys()):
    tem_data=c[i]
    for j in range(len(tem_data)):
        code_list.loc[n]=[tem_data[j].ccode,tem_data[j].info,tem_data[j].market,tem_data[j].name]
        n+=1
print(code_list)
```

```python
all_list=pd.DataFrame({"value":[],"eps":[],"bps":[],"dividennd_one":[]})

n=0

for i in tqdm(code_list["ccode"]):
    try:
        df_f=q.get_finance(i)
        df_v=q.get_price(i)
        all_list.loc[n]=[df_f.bps,df_f.dividend_one,df_f.eps,df_v.close]
        n+=1
    except:
        all_list.loc[n]=[None,None,None,None]
        n+=1
        pass
```

```python
all_list["name"]=code_list["name"]
all_list["ccode"]=code_list["ccode"]
all_list=all_list.dropna(how="any")
all_list.head()

all_list.to_csv(expanduser("~")+"/Desktop/stock/"+str(now)+".csv")
```
