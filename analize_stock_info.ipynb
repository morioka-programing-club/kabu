```python
import pandas as pd
import numpy as np
import datetime
from tqdm import tqdm
import matplotlib.pyplot as plt
import sklearn
from sklearn import linear_model
from os.path import expanduser

clf = linear_model.LinearRegression(fit_intercept=True, normalize=True, copy_X=True)

now = datetime.date.today()
```

```python
stock_data=pd.read_csv(filepath_or_buffer=expanduser("~")+"/Desktop/stock/"+str(now)+".csv")#, usecols=["bps","eps","dividennd_one","value","name","ccode"])

stock_data.head()
```

```python
sklearn.linear_model.LinearRegression(fit_intercept=True, normalize=True,copy_X=True, n_jobs=1)
```

```python
X =stock_data[["bps","dividennd_one","eps"]]
Y =stock_data["value"]

clf.fit(X,Y)
print(clf.coef_)
stock_data["predict"]=clf.predict(X)
stock_data["wariyasudo"]=stock_data["predict"]/stock_data["value"]
stock_data=stock_data.sort_values("wariyasudo", ascending=False)
print(stock_data)
```

```python
stock_data.to_csv("C:/Users/User/Desktop/stock/"+str(now)+".result.csv", encoding='utf_8_sig')
```
